```python
import asyncio
import random
from playwright.async_api import async_playwright

# Función para simular un comportamiento humano con tiempos de espera aleatorios
async def simular_comportamiento_humano(page):
    await asyncio.sleep(random.uniform(2, 5))  # Espera aleatoria entre 2 y 5 segundos
    await page.mouse.move(random.uniform(0, 800), random.uniform(0, 600))
    await asyncio.sleep(random.uniform(1, 3))  # Otra espera aleatoria

async def close_popup(page):
    try:
        await page.keyboard.press('Escape')
        await page.wait_for_timeout(1000)
        await page.keyboard.press('Escape')
        await asyncio.sleep(random.uniform(1, 3))
        await page.keyboard.press('Escape')
        await asyncio.sleep(random.uniform(1, 3))
        print("Popup cerrado usando la tecla Escape.")
    except Exception as e:
        print(f"No se pudo cerrar el popup con la tecla Escape: {e}")

async def determine_page_metrics(page):
    """Determina las métricas de la página: altura del viewport y tamaño del documento."""
    try:
        viewport_height = await page.evaluate('window.innerHeight')
        document_height = await page.evaluate('document.documentElement.scrollHeight')
        return viewport_height, document_height
    except Exception as e:
        print(f"No se pudieron obtener las métricas de la página: {e}")
        return None, None

async def scroll_slowly_to_end(page, scroll_percentage):
    """Realiza scroll lentamente hasta que se alcanza un porcentaje específico de la página."""
    try:
        viewport_height, document_height = await determine_page_metrics(page)
        if not viewport_height or not document_height:
            print("No se pudieron determinar las métricas de la página.")
            return False

        scroll_target = document_height * scroll_percentage
        print(f"Scroll target: {scroll_target}")

        scroll_pause_time = random.uniform(0.3, 1)  # Tiempo entre cada paso de scroll (ajustable) ## Los parámetros con los que fijo funciona random.uniform(0.5, 2)
        scroll_amount = random.randint(80, 120)  # Cantidad de pasos de scroll (ajustable) ## Los parámetros con los que fijo funciona random.randint(100, 180)

        for _ in range(scroll_amount):
            await page.evaluate('window.scrollBy(0, window.innerHeight * 0.1)')  # Scroll hacia abajo en un 10% del viewport ## Los parámetros con los que fijo funciona 'window.scrollBy(0, window.innerHeight * 0.05)'
            await page.wait_for_timeout(scroll_pause_time * 100)  # Esperar entre cada paso de scroll

            # Movimiento del ratón opcional  ## En caso de no funcionar el código o ser detectado se puede disminuir el valor del 0.65
            if random.random() > 0.65:  # Solo mover el ratón ocasionalmente
                mouse_x = random.randint(0, viewport_height)
                mouse_y = random.randint(0, viewport_height)
                await page.mouse.move(mouse_x, mouse_y, steps=5)

            current_scroll_position = await page.evaluate('window.scrollY')
            if current_scroll_position >= scroll_target:
                break

            # Reevaluar las métricas
            new_viewport_height, new_document_height = await determine_page_metrics(page)
            if new_document_height != document_height:
                document_height = new_document_height
                scroll_target = document_height * scroll_percentage
                print(f"Nuevo scroll target: {scroll_target}")

            #Realizar un close popup para cerrar publicada emergente continua ocasionalmente
            if _ % 80 == 0:
              await close_popup(page)

        print(f"Scroll lento completado hasta el {scroll_percentage * 100}% de la página.")
        await close_popup(page)

        return True
    except Exception as e:
        print(f"No se pudo completar el scroll lento: {e}")
        return False

# Función mejorada para extraer datos de estancias
async def run_extraccion_data_estancias(playwright, url, browser, user_agent):
    lista_id_estancias_func = []
    lista_nombre_inicial_estancia_func = []
    lista_cantidad_id_x_pagina_func = []
    lista_urls_id_func = []
    lista_precios_sin_iva = []
    lista_descripcion_estancia = []
    lista_ubicacion_estancia = []

    # Crear un nuevo contexto de navegador con un user-agent personalizado
    context = await browser.new_context(
        user_agent=user_agent,
        viewport={"width": random.randint(1200, 1920), "height": random.randint(800, 1080)},  # Resolución aleatoria
        device_scale_factor=random.uniform(1, 2),  # Factor de escala aleatorio
        accept_downloads=False,
        locale="es-ES",  # Localización en español para parecer más legítimo
    )

    page = await context.new_page()

    try:
        # Navegar a la URL especificada
        await page.goto(url, timeout=90000)

        # Simular comportamiento humano para evitar ser detectado
        await simular_comportamiento_humano(page)

        # Esperar que la página se cargue completamente
        await page.wait_for_load_state("networkidle", timeout=90000)
        await asyncio.sleep(random.uniform(2, 5))

        # Intentar encontrar el contenedor principal usando XPath o CSS
        container1 = await page.query_selector('div[data-results-container="1"]')
        container2 = await page.query_selector('//*[@id="bodyconstraint-inner"]/div[2]/div/div[2]/div[3]/div[2]/div[2]/div[3]')
        container = container1 or container2

        if container:
            await close_popup(page)
            divs = await container.query_selector_all('div[data-testid="property-card-container"]')
            hrefs = await container.query_selector_all('a[data-testid="title-link"]')
            contenedor_calen1 = await page.query_selector('div[data-testid="searchbox-dates-container"]')
            contenedor_calen2 = await page.query_selector('//*[@id="bodyconstraint-inner"]/div[2]/div/div[1]/div/form/div[1]/div[2]/div/div')
            contenedor_calen3 = await page.query_selector('#bodyconstraint-inner > div:nth-child(8) > div > div.a6e267116d > div > form > div.ffb9c3d6a3.b3b8f00b52.c9a7790c31.e691439f9a > div:nth-child(2) > div > div')
            contenedor_calen = contenedor_calen1 or contenedor_calen2 or contenedor_calen3
            if contenedor_calen:
                await page.screenshot(path="imagen_inicio_navegacion.png")
                print(f"Se encontraron elementos <div> con el atributo data-testid='searchbox-dates-container' dentro del contenedor especificado.")
                #await contenedor_calen.click()
                await asyncio.sleep(random.uniform(2, 5))
                await page.screenshot(path="imagen_click_calen.png")
            else:
                print("No se encontraron elementos <div> con el atributo data-testid='searchbox-dates-container' dentro del contenedor especificado.")

            if divs:
                print(f'Se encontraron {len(divs)} elementos <div> con el atributo data-testid="property-card-container" dentro del contenedor especificado.')
                print(f"Se encontraron {len(hrefs)} elementos <href> con el atributo data-testid='title-link' dentro del contenedor especificado.")

                div_calen = await page.query_selector('div[data-testid="searchbox-datepicker"]')

                if div_calen:
                    print(f"Se encontraron elementos <div> con el atributo data-testid='searchbox-datepicker' dentro del contenedor especificado.")
                    #texto_html_cal = await div_calen.inner_html()
                    #print(texto_html_cal)

                    # 2. OBTENER LAS FECHAS ACTIVAS DEL INMUEBLE

                    from datetime import datetime

                    # Función para procesar los valores antes de añadirlos a la lista
                    def procesar_fecha(fecha):
                        if pd.isna(fecha):  # Verifica si la fecha es NaT
                            return None  # O usa 'No Date' si prefieres un texto específico
                        return fecha

                    async def fun_obtener_fechas_activas(contenedor_fechas):
                      selector_fechas_activas = '//span[not(contains(@class, "fa1e2a2fae")) and @role="checkbox"]'
                      fechas_activas_element = await contenedor_fechas.query_selector_all(selector_fechas_activas)
                      lista_fechas_activas = []
                      if fechas_activas_element:
                          for data_fecha in fechas_activas_element:
                            # print(f'El elemento data_fecha es: {data_fecha}')
                            try:
                              elemento_fecha = await data_fecha.get_attribute('data-date') # Extrae los datos de si la fecha es válida y si es para check in o check out
                              contenido_div_fecha = await data_fecha.get_attribute('aria-label')
                              if elemento_fecha and contenido_div_fecha:
                                lista_fechas_activas.append((contenido_div_fecha,elemento_fecha)) # Se crea texto con los datos de las fechas
                              else:
                                print('No se encontró el elemento de fecha')
                                lista_fechas_activas.append((procesar_fecha(pd.NaT),'No hay fechas elegidas'))

                            except Exception as e:
                              print(f"Ocurrió un error: {e}")
                              lista_fechas_activas.append((procesar_fecha(pd.NaT),'No hay fechas elegidas'))
                              pass
                      else:
                          fechas_activas_element = 'None'
                          print(f"No se encontró el elemento de fechas activas {selector_fechas_activas}")
                          lista_fechas_activas.append((procesar_fecha(pd.NaT),'No hay fechas elegidas'))

                      #print(f"Las fechas activas son: {lista_fechas_activas}")
                      return lista_fechas_activas

                    lista_fechas_entrada_activas =  await fun_obtener_fechas_activas(div_calen)
                    print(f"Las fechas activas son: {lista_fechas_entrada_activas}")

                    # Si existen fechas activas
                    if lista_fechas_entrada_activas:
                      try:
                        # 3. SELECCIONAR LA PRIMERA FECHA
                        if len(lista_fechas_entrada_activas) > 2:
                          index_fecha_elegida = random.randint(0, len(lista_fechas_entrada_activas) - 2)
                          selector_primera_fecha_activa = f'span[data-date="{lista_fechas_entrada_activas[index_fecha_elegida][1]}"]'
                          elements_click_fecha1 = await page.query_selector(selector_primera_fecha_activa)
                          print(f'El elemento con el selector {selector_primera_fecha_activa} es: {elements_click_fecha1}')
                          await elements_click_fecha1.click()
                          await page.screenshot(path="imagen_click_fecha1.png")
                          selector_primera_fecha_activa2 = f'span[data-date="{lista_fechas_entrada_activas[index_fecha_elegida+1][1]}"]'
                          elements_click_fecha2 = await page.query_selector(selector_primera_fecha_activa2)
                          print(f'El elemento con el selector {selector_primera_fecha_activa2} es: {elements_click_fecha2}')
                          await elements_click_fecha2.click()
                          await page.screenshot(path="imagen_click_fecha2.png")
                        elif len(lista_fechas_entrada_activas) == 2:
                          selector_primera_fecha_activa = f'span[data-date="{lista_fechas_entrada_activas[index_fecha_elegida][0][1]}"]'
                          elements_click_fecha1 = await page.query_selector(selector_primera_fecha_activa)
                          print(f'El elemento con el selector {selector_primera_fecha_activa} es: {elements_click_fecha1}')
                          await elements_click_fecha1.click()
                          await page.screenshot(path="imagen_click_fecha1.png")
                          selector_primera_fecha_activa2 = f'span[data-date="{lista_fechas_entrada_activas[index_fecha_elegida][1][1]}"]'
                          elements_click_fecha2 = await page.query_selector(selector_primera_fecha_activa2)
                          print(f'El elemento con el selector {selector_primera_fecha_activa2} es: {elements_click_fecha2}')
                          await elements_click_fecha2.click()
                          await page.screenshot(path="imagen_click_fecha2.png")
                        else:
                          print("No hay fechas activas disponibles.")

                      except:
                        print(f"No se encontró el elemento con el selector_primera_fecha_activa")
                        #VALIDAR SI SE PONE UN PASS EN EL EXCEPT PARA QUE EL PROCESO SIGA DADO QUE LOS PRECIOS SE PUEDEN OBTENER ESTANDO DENTRO DE CADA PUBLICACIÓN

                      #BUSCAR Y DAR CLICK SOBRE EL BOTÓN BUSCAR
                      selector_boton_buscar = 'button[type="submit"]'
                      elements_click_boton_buscar = await page.query_selector(selector_boton_buscar)
                      if elements_click_boton_buscar:
                        print(f'El elemento con el selector {selector_boton_buscar} es: {elements_click_boton_buscar}')
                        await elements_click_boton_buscar.click()
                        await page.screenshot(path="imagen_click_boton_buscar.png")
                      else:
                        print(f"No se encontró el elemento con el selector {selector_boton_buscar}")
                else:
                    print("No se encontraron elementos <div> con el atributo data-testid='searchbox-datepicker' dentro del contenedor especificado.")

                await asyncio.sleep(random.uniform(2, 5))

                # GUARDAR EN VARIABLE EL NÚMERO TOTAL DE RESULTADOS ESPERADOS

                select_cantidad_resultados_esperados = await page.query_selector('h1[aria-live="assertive"]')
                if select_cantidad_resultados_esperados:
                  texto_cantidad_resultados_esperados = await select_cantidad_resultados_esperados.inner_text()
                  num_matches = int(''.join(re.findall(r'\d+', texto_cantidad_resultados_esperados)))
                  print(f'El número de resultados esperados es: {num_matches}')

                # PASOS PARA EXTRAER INFORMACIÓN ASOCIADA AL PRECIOS Y LA DESCRIPCIÓN /// SE CREA FUNCIÓN PARA EXTRACCIÓN DE DATA ESTOY PENSANDO EN HACER UN CICLO WHILE PARA EXTRAER LA INFOR MIETRAS EXISTE EL BOTÓN DE MÁS ANUNCIONS
                async def funcion_extraer_info_malla():
                    divs = await page.query_selector_all('div[data-testid="property-card-container"]')
                    if not divs:
                        print("No se encontraron elementos <div> con el atributo data-testid='property-card-container' dentro del contenedor especificado.")
                    else:
                        for div in divs:
                            try:
                                # Buscar la URL de la estancia
                                href = await div.query_selector('a[data-testid="title-link"]')
                                if href:
                                    href_estancia = await href.get_attribute('href')
                                    id_estancia = href_estancia.split('?')[0].split('/')[5]
                                    if id_estancia in lista_id_estancias_func:
                                        continue
                                    else:
                                      lista_urls_id_func.append(href_estancia)
                                      lista_id_estancias_func.append(id_estancia)
                                else:
                                    print("No se encontró el elemento <a> con el atributo data-testid='title-link'.")
                                    lista_urls_id_func.append('No hay URL')
                                    lista_id_estancias_func.append('No ID estancia')

                                # Buscar el nombre de la estancia
                                nombre_estancia = await div.query_selector('h3')
                                if nombre_estancia:
                                    texto_nombre_estancia = await nombre_estancia.inner_text()
                                    lista_nombre_inicial_estancia_func.append(texto_nombre_estancia)
                                else:
                                    print("No se encontró el elemento <div> con el atributo data-testid='title'.")
                                    lista_nombre_inicial_estancia_func.append('No hay nombre de la estancia')

                                # Realizar la búsqueda de la comuna o ubicación general
                                comuna_general = await div.query_selector('span[data-testid="address"]')
                                if comuna_general:
                                    texto_comuna_general = await comuna_general.inner_text()
                                    lista_ubicacion_estancia.append(texto_comuna_general)
                                else:
                                    print("No se encontró el elemento <span> con el atributo data-testid='address'.")
                                    lista_ubicacion_estancia.append('No hay datos de comuna o localización general')

                                # Repetir la búsqueda del contexto de la estancia
                                contexto_estancia = await div.query_selector('div[data-testid="recommended-units"]')

                                if contexto_estancia:
                                    texto_descripcion_estancia = await contexto_estancia.inner_text()
                                    lista_descripcion_estancia.append(texto_descripcion_estancia)
                                    #print(f"Se encontraron elementos <div> con el atributo data-testid='recommended-units' dentro del contenedor especificado.")
                                else:
                                    print("No se encontraron elementos <div> con el atributo data-testid='recommended-units'.")
                                    lista_descripcion_estancia.append('No hay datos generales')

                                # Buscar elementos de precios sin IVA en el contenedor actual y la página completa
                                precio_sin_iva1 = await div.query_selector_all('div[data-testid="availability-rate-information"]')
                                precio_sin_iva2 = await page.query_selector_all('div[data-testid="availability-rate-information"]')
                                precio_sin_iva = precio_sin_iva1 if precio_sin_iva1 else precio_sin_iva2

                                if precio_sin_iva:
                                    #print(f"Se encontraron {len(precio_sin_iva)} elementos <div> con el atributo de availability-rate-information.")
                                    for precio in precio_sin_iva:
                                        previo_valor_noche = await precio.inner_text()
                                        if previo_valor_noche is not None:
                                            lista_precios_sin_iva.append(previo_valor_noche)
                                        else:
                                            lista_precios_sin_iva.append('No se encontraron precios')
                                else:
                                    print("No se encontraron elementos <div> con el atributo de availability-rate-information.")
                                    lista_precios_sin_iva.append('No se encontraron precios')

                            except Exception as e:
                                print(f"Error en la extracción: {str(e)}")
                                lista_descripcion_estancia.append('No hay datos generales')
                                lista_precios_sin_iva.append('No se encontraron precios')

                        lista_cantidad_id_x_pagina_func.append(len(lista_id_estancias_func))
                        print(f'La cantidad de ID estancias unicas capturadas por ahora es: {len(lista_id_estancias_func)}')

                #PASOS PARA VALIDAR SI EL BOTÓN "Cargar más resultados" ESTÁ VISIBLE AL TERMINAR EL SCROLL LENTO

                async def encontrar_boton_mas_anuncios(page):
                    """Busca el botón de 'Cargar más resultados' usando diferentes selectores."""
                    boton_selector1 = await page.query_selector('button[class="a83ed08757 c21c56c305 bf0537ecb5 f671049264 af7297d90d c0e0affd09"]')
                    boton_selector2 = await page.query_selector('//span[contains(text(), "Cargar más resultados") or contains(text(), "Load more results")]')
                    return boton_selector1 or boton_selector2

                async def scroll_and_wait(page, contador_scroll, filename_prefix):
                    """Realiza el scroll lentamente hasta un porcentaje aleatorio de la página y toma una captura."""
                    scroll_percentage = random.uniform(0.85, 0.95)
                    await scroll_slowly_to_end(page, scroll_percentage)  # Asumo que esta función está definida
                    #await page.screenshot(path=f"{filename_prefix}_{contador_scroll}.png")
                    await asyncio.sleep(random.uniform(2, 5))

                async def click_boton_mas_anuncios(page, boton, contador_scroll):
                    """Intenta hacer clic en el botón de 'Cargar más resultados' y espera la carga de nuevos elementos."""
                    try:
                        print(f'Clic en el botón de "Cargar más resultados" (Intento: {contador_scroll})')
                        await boton.click()
                        await asyncio.sleep(random.uniform(5, 10))  # Espera después del clic para cargar resultados
                        #await page.screenshot(path=f"imagen_click_boton_mas_anuncios_{contador_scroll}.png")
                    except Exception as e:
                        print(f"Error al hacer clic en el botón de 'Cargar más resultados': {e}")

                async def extraer_mas_anuncios(page):
                    """Proceso completo para realizar scrolls y cargar más resultados hasta que ya no haya más."""
                    contador_scroll = 0
                    boton_mas_anuncios = await encontrar_boton_mas_anuncios(page)

                    # Primer ciclo: Scroll lento hasta encontrar el botón
                    while not boton_mas_anuncios:
                        contador_scroll += 1
                        print('Scroll hasta encontrar el botón de más anuncios')
                        await scroll_and_wait(page, contador_scroll, "imagen_scroll")
                        boton_mas_anuncios = await encontrar_boton_mas_anuncios(page)
                        # Tiempo de espera para la carga de los nuevos resultados
                        await asyncio.sleep(random.uniform(2, 5))
                        await funcion_extraer_info_malla()

                    # Segundo ciclo: Cargar más anuncios haciendo clic en el botón de "Cargar más resultados"
                    contador_scroll2 = 0
                    while boton_mas_anuncios:
                        contador_scroll2 += 1
                        await click_boton_mas_anuncios(page, boton_mas_anuncios, contador_scroll2)
                        print('Scroll después de hacer clic en el botón de más anuncios')
                        await scroll_and_wait(page, contador_scroll2, "imagen_scroll_dentro_boton")
                        boton_mas_anuncios = await encontrar_boton_mas_anuncios(page)
                        # Tiempo de espera para la carga de los nuevos resultados
                        await asyncio.sleep(random.uniform(2, 5))
                        await funcion_extraer_info_malla()

                    print('No se encontraron más botones de "Cargar más resultados". Proceso completado.')

                # Llamada principal al proceso de extracción
                await extraer_mas_anuncios(page)

                #await page.screenshot(path="imagen_click_boton_mas_anuncios2.png")
                # await funcion_extraer_info_malla()
                await asyncio.sleep(random.uniform(2, 5))

                # Finalmente, imprimimos los resultados de las listas
                #print(f'La lista de URLs de las estancias es: {lista_urls_id_func}')
                #print(f'La lista de descripciones de estancia es: {lista_descripcion_estancia}')
                #print(f'La lista de precios sin IVA es: {lista_precios_sin_iva}')
                #print(f'La lista de ubicaciones de las estancias es: {lista_ubicacion_estancia}')

            else:
                lista_id_estancias_func.append('No ID de estancia encontrado')
                lista_nombre_inicial_estancia_func.append('No se encontró el nombre de la estancia')
                lista_urls_id_func.append('No URL encontrada')
                lista_precios_sin_iva.append('No se encontraron precios sin IVA')
                lista_cantidad_id_x_pagina_func.append('No IDs encontrados en esta página')
                lista_descripcion_estancia.append('No hay descripción disponible')
                lista_ubicacion_estancia.append('No se encontró ubicación')
                print("No se encontraron elementos <div> con el atributo data-testid='listing-card-title' dentro del contenedor especificado.")
        else:
            lista_id_estancias_func.append('No ID de estancia encontrado')
            lista_nombre_inicial_estancia_func.append('No se encontró el nombre de la estancia')
            lista_urls_id_func.append('No URL encontrada')
            lista_precios_sin_iva.append('No se encontraron precios sin IVA')
            lista_cantidad_id_x_pagina_func.append('No IDs encontrados en esta página')
            lista_descripcion_estancia.append('No hay descripción disponible')
            lista_ubicacion_estancia.append('No se encontró ubicación')
            print("No se encontró el contenedor especificado.")

    except Exception as e:
        lista_id_estancias_func.append('No ID de estancia encontrado')
        lista_nombre_inicial_estancia_func.append('No se encontró el nombre de la estancia')
        lista_urls_id_func.append('No URL encontrada')
        lista_precios_sin_iva.append('No se encontraron precios sin IVA')
        lista_cantidad_id_x_pagina_func.append('No IDs encontrados en esta página')
        lista_descripcion_estancia.append('No hay descripción disponible')
        lista_ubicacion_estancia.append('No se encontró ubicación')
        print(f"Error en la extracción: {e}")

    finally:
        await browser.close()  # Asegurarse de cerrar el navegador
        return lista_id_estancias_func, lista_nombre_inicial_estancia_func, lista_cantidad_id_x_pagina_func, lista_urls_id_func, lista_precios_sin_iva, lista_descripcion_estancia, lista_ubicacion_estancia

def ajustar_longitudes_iguales(*listas):
    longitud_maxima = max(len(lista) for lista in listas)
    for lista in listas:
        while len(lista) < longitud_maxima:
            lista.append(0)  # Puedes cambiar `None` por un valor predeterminado según lo que necesites

# Función principal con configuración de user-agents y rotación de navegadores
async def main(url, lista_id_estancias, lista_nombre_inicial_estancia, lista_cantidad_id_x_pagina, lista_urls_id, lista_precios_sin_iva_, lista_descripcion_estancia, lista_ubicacion_estancia):

    channels = ['chromium', 'msedge', 'chrome']  # Canales de Chromium
    async with async_playwright() as p:
        channel = random.choice(channels)
        user_agent = random.choice(user_agents)
        try:
            await asyncio.sleep(2)
            browser = await p.chromium.launch(channel=channel, headless=True)
            extracto = await run_extraccion_data_estancias(p, url, browser, user_agent)
            lista_id_estancias.append(extracto[0])
            lista_nombre_inicial_estancia.append(extracto[1])
            lista_cantidad_id_x_pagina.append(extracto[2])
            lista_urls_id.append(extracto[3])
            lista_precios_sin_iva_.append(extracto[4])
            lista_descripcion_estancia.append(extracto[5])
            lista_ubicacion_estancia.append(extracto[6])
            print(f"Resultados obtenidos con {channel} y user-agent {user_agent}: {extracto}")
        except Exception as e:
            print(f"Error en {channel} con user-agent {user_agent}: {e}")
        finally:
            await browser.close()
            await asyncio.sleep(2)
        # Ajustar todas las listas a la misma longitud
        ajustar_longitudes_iguales(lista_id_estancias, lista_nombre_inicial_estancia, lista_cantidad_id_x_pagina, lista_urls_id, lista_precios_sin_iva_, lista_descripcion_estancia, lista_ubicacion_estancia)

        return lista_id_estancias, lista_nombre_inicial_estancia, lista_cantidad_id_x_pagina, lista_urls_id, lista_precios_sin_iva_, lista_descripcion_estancia, lista_ubicacion_estancia
```

```python
# Definir una función que crea un DataFrame con los datos de estancias de booking
def def_crear_df_id_estancias(lista_id_estancias, lista_nombre_inicial_estancia, lista_urls_id, lista_precios_sin_iva_, lista_descripcion_estancia, lista_ubicacion_estancia):

    # Crear un DataFrame con los datos combinados de varias listas usando itertools.chain para "aplanar" las listas anidadas
    data_frame_booking = pd.DataFrame({'id': list(itertools.chain(*lista_id_estancias)),
                                      'nombre': list(itertools.chain(*lista_nombre_inicial_estancia)),
                                      'url': list(itertools.chain(*lista_urls_id)),
                                      'precio_publicado': list(itertools.chain(*lista_precios_sin_iva_)),
                                      'informacion_general': list(itertools.chain(*lista_descripcion_estancia)),
                                      'localizacion_general': list(itertools.chain(*lista_ubicacion_estancia))})

    # Eliminar duplicados en el DataFrame basado en todas las columnas
    data_frame_booking = data_frame_booking.drop_duplicates()

    # Extraer el nombre real de la estancia
    data_frame_booking['nombre'] = data_frame_booking['nombre'].apply(lambda x: x.split('\n')[0].strip())

    # Crear una nueva columna 'link_booking' que contiene un enlace a cada estancia basado en el 'id'
    data_frame_booking.loc[:, 'link_booking'] = data_frame_booking['id'].apply(lambda x: f"https://www.booking.com/hotel/co/{x}")

    # Aplicar una función personalizada (extraer_precios) para extraer los precios y la moneda de la columna 'precio_publicado'
    # Se dividen los valores devueltos en tres nuevas columnas: 'precio_actual_x_noche', 'precio_original_x_noche' y 'moneda'
    data_frame_booking[['noches', 'adultos', 'precio_actual', 'precio_original', 'moneda']] = data_frame_booking['precio_publicado'].apply(
        lambda x: pd.Series(extraer_precios(x)))

    # Convertir las columnas 'precio_actual_x_noche' y 'precio_original_x_noche' a tipo float
    data_frame_booking['precio_actual_x_noche'] = data_frame_booking['precio_actual']/ data_frame_booking['noches']
    data_frame_booking['precio_original_x_noche'] = data_frame_booking['precio_original']/ data_frame_booking['noches']

    # Crear una nueva columna 'precios_diferentes' que verifica si el precio actual es diferente al precio original
    data_frame_booking['precios_diferentes'] = np.array(data_frame_booking['precio_actual_x_noche']) != np.array(data_frame_booking['precio_original_x_noche'])

    # Agregar una columna 'dia_extraccion' que almacena la fecha actual en formato 'YYYY-MM-DD'
    data_frame_booking['dia_extraccion'] = pd.to_datetime('today').strftime('%Y-%m-%d')

    # Filtrar las filas donde 'id' no es válido y resetear los índices del DataFrame
    data_frame_booking = data_frame_booking[~data_frame_booking['id'].isin(['No ID estancia','No ID de estancia encontrado'])].reset_index(drop=True)

    # Ordenar el DataFrame por 'id', 'precio_actual_x_noche' y 'dia_extraccion' en orden descendente
    data_frame_booking = data_frame_booking.sort_values(by=['id', 'precio_actual_x_noche', 'dia_extraccion'], ascending=(False, False, False)).reset_index(drop=True)

    # Crear un nuevo DataFrame 'data_frame_booking_unicos_recientes' eliminando duplicados basados en la columna 'id'
    # Mantener solo la fila más reciente para cada 'id'
    data_frame_booking_unicos_recientes = data_frame_booking.drop_duplicates(subset=['id']).reset_index(drop=True)

    #Data frame simplificado
    data_frame_booking_unicos_recientes = data_frame_booking_unicos_recientes.drop(columns=['informacion_general','localizacion_general'])

    # Definir los patrones problemáticos y sus reemplazos
    caracteres_problematicos = {
        r'\t': '|',   # Reemplaza tabuladores con un espacio
        r'\n': '|',   # Reemplaza saltos de línea con un espacio
        r'\r': '|'    # Reemplaza retornos de carro con un espacio
    }

    #data_frame_booking_unicos_recientes = data_frame_booking_unicos_recientes.replace(caracteres_problematicos, regex=True)
    data_frame_booking = data_frame_booking.replace(caracteres_problematicos, regex=True)

    # Crear un nuevo DataFrame 'data_frame_booking_unicos_recientes' eliminando duplicados basados en la columna 'id'
    # Mantener solo la fila más reciente para cada 'id', ''dia_extraccion','precio_publicado'
    data_frame_booking_unicos_id_precio_fecha_extraccion = data_frame_booking.drop_duplicates(subset=['id', 'dia_extraccion','precio_publicado']).reset_index(drop=True)

    # Obtener las monedas únicas del DataFrame de estancias
    monedas_base = data_frame_booking[data_frame_booking['moneda']!='COP']['moneda'].unique()

    # Obtener las tasas de cambio para todas las monedas únicas hacia COP
    tasa_monedas_to_cop = obtener_tasa_cambio(monedas_base, 'COP')

    # Solo imprime las tasas si hay nuevos datos
    if tasa_monedas_to_cop is not None:
        print(tasa_monedas_to_cop)

    # Retornar ambos DataFrames: el original procesado y el que contiene solo las filas únicas más recientes
    return data_frame_booking, data_frame_booking_unicos_recientes, data_frame_booking_unicos_id_precio_fecha_extraccion

# Función para extraer precios y posibles monedas
def extraer_precios(precio):
    # Lista ampliada de posibles monedas y sus símbolos/abreviaturas
    monedas = {
        'USD': ['$', 'usd', 'dólar', 'dolares', 'dollars'],
        'EUR': ['€', 'eur', 'euro', 'euros'],
        'GBP': ['£', 'gbp', 'libra', 'libras', 'pound'],
        'TWD': ['twd', 'nt$', 'new taiwan dollar'],
        'JPY': ['¥', 'jpy', 'yen', 'yens'],
        'CAD': ['cad', 'can$', 'canadian dollar'],
        'AUD': ['aud', 'a$', 'australian dollar'],
        'MXN': ['mxn', 'mx$', 'peso mexicano', 'pesos mexicanos'],
        'BRL': ['brl', 'r$', 'real', 'reais'],
        'ARS': ['ars', 'ar$', 'peso argentino', 'pesos argentinos'],
        'COP': ['cop', 'col$', 'peso colombiano', 'pesos colombianos'],
        'CLP': ['clp', 'cl$', 'peso chileno', 'pesos chilenos'],
        'PEN': ['pen', 's/.', 'sol', 'soles'],
        'UYU': ['uyu', 'uy$', 'peso uruguayo', 'pesos uruguayos'],
        'PYG': ['pyg', 'gs.', 'guaraní', 'guaraníes'],
        'VES': ['ves', 'bs.', 'bolívar', 'bolívares'],
        'CRC': ['crc', '₡', 'colón', 'colones'],
        'PAB': ['pab', 'b/.', 'balboa', 'balboas'],
        'DOP': ['dop', 'rd$', 'peso dominicano', 'pesos dominicanos'],
        'BOB': ['bob', 'bs.', 'boliviano', 'bolivianos'],
        'CUP': ['cup', 'cup$', 'peso cubano', 'pesos cubanos'],
        'GTQ': ['gtq', 'q', 'quetzal', 'quetzales'],
        'HNL': ['hnl', 'lempira', 'lempiras'],
        'CNY': ['cny', 'rmb', 'yuan', 'renminbi', '¥'],
        'KRW': ['krw', '₩', 'won'],
        'INR': ['inr', '₹', 'rupee', 'rupia', 'rupias'],
        'RUB': ['rub', '₽', 'ruble', 'rublo', 'rublos'],
        # Agrega más monedas si es necesario
    }

    # Convertir todo a minúsculas
    noches = int(re.findall(r'\d+', str(precio).lower().split(',',1)[0].strip())[0])
    noches = 1 if noches == 0 else noches
    adultos = int(re.findall(r'\d+', str(precio).lower().split(',',1)[1].replace('\xa0', ' ').strip().split('\n')[0])[0])
    precio_limpio =  ''.join(str(precio).lower().split(',',1)[1].replace('\xa0', ' ').strip().split('\n')[1])
    #print(f'El precio limpio es: {precio_limpio}')

    # Buscar símbolos de monedas en el texto
    moneda_detectada = 'USD'  # Valor predeterminado si no se encuentra otro símbolo de moneda

    for moneda, simbolos in monedas.items():
        for simbolo in simbolos:
            if simbolo in precio_limpio:
                moneda_detectada = moneda
                break
        if moneda_detectada != 'USD':  # Si se encontró una moneda, salir del bucle
            break

    def convertir_a_float(numero_str):
        """
        Convierte una cadena numérica a float, distinguiendo entre separador de miles y decimales.
        """
        # Si el número contiene tanto ',' como '.', vamos a suponer que ',' es separador de miles y '.' es decimal
        if ',' in numero_str and '.' in numero_str:
            # Eliminamos la coma como separador de miles
            return float(numero_str.replace(',', ''))

        # Si solo contiene coma, es posible que sea separador decimal
        elif ',' in numero_str:
            # Si hay exactamente tres dígitos después de la coma, es un separador de miles
            partes = numero_str.split(',')
            if len(partes[-1]) == 3:
                # Eliminar la coma (es separador de miles)
                return float(numero_str.replace(',', ''))
            else:
                # Reemplazar la coma por un punto (es decimal)
                return float(numero_str.replace(',', '.'))

        # Si solo contiene punto, tratamos de determinar si es separador de miles o decimal
        elif '.' in numero_str:
            partes = numero_str.split('.')
            if len(partes[-1]) == 3:
                # Eliminar el punto (es separador de miles)
                return float(numero_str.replace('.', ''))
            else:
                # Mantener el punto (es decimal)
                return float(numero_str)

        # Si no contiene ni punto ni coma, lo convertimos directamente a float
        return float(numero_str)

    def extraer_y_procesar_precios(precio_limpio):
        # Expresión regular para capturar los números
        match = re.findall(r'\d+(?:[.,]\d+)*', precio_limpio)
        #print(f'Los precios encontrados son: {match}')

        if len(match) == 2:
            # Convertir ambos números a float usando la función de conversión
            num1 = convertir_a_float(match[0])
            num2 = convertir_a_float(match[1])

            # Obtener el valor máximo y mínimo
            max_value = max(num1, num2)
            min_value = min(num1, num2)

            #print(f"Max: {max_value}, Min: {min_value}")
            return noches, adultos, min_value, max_value, moneda_detectada
        elif len(match) == 1:
            num1 = convertir_a_float(match[0])
            return noches, adultos, num1, num1, moneda_detectada
        else:
            return 'No se encuentran noches', 'No se identifican adultos ',0, 0, moneda_detectada


    return extraer_y_procesar_precios(precio_limpio)

# Definir la ruta donde se guardarán los archivos CSV del histórico
ruta_historico = "/content/drive/MyDrive/LABORAL/ALCALDÍA/HACIENDA/WEB_SCRAPING/AIRBNB/HISTORIAL_PRECIOS/historico_tasas_cambio.csv"

# Función para verificar si es un nuevo día
def es_nuevo_dia(monedas_base, ruta_historico = ruta_historico):
    fecha_actual = pd.to_datetime('today').strftime('%Y-%m-%d')

    # Si el archivo no existe, es un nuevo día (primera vez)
    if not os.path.exists(ruta_historico):
        print('no existe')
        return True

    # Leer la última fecha guardada
    fecha_pd = pd.read_csv(ruta_historico,sep='|',encoding='utf-8-sig')
    #display(fecha_pd)
    ultima_fecha = pd.to_datetime(fecha_pd['fecha'].iloc[-1]).strftime('%Y-%m-%d')
    lista_tipo_moneda = set(fecha_pd[fecha_pd['moneda'] != 'COP']['moneda'].unique())

    # Comparar la última fecha guardada con la fecha actual
    if (fecha_actual != ultima_fecha) or (set(monedas_base) != lista_tipo_moneda):
        # Si es un nuevo día, actualizar el archivo con la nueva fecha
        return True
    else:
        return False

# Función para obtener la tasa de cambio para varias monedas base a una moneda objetivo
def obtener_tasa_cambio(monedas_base, target):
    lista_moneda_base = []
    lista_valor_conversion = []
    lista_fecha = []
    # Lista de nuevas monedas a añadir
    nuevas_monedas = ['EUR', 'USD', 'CAD', 'JPY','TWD']

    # Convertir la lista a un array de NumPy y combinarlo con el array original
    monedas_base = np.append(monedas_base, nuevas_monedas)

    # Eliminar duplicados garantizando que los valores sean únicos
    monedas_base = np.unique(monedas_base)

    # Solo ejecuta si es un nuevo día
    if es_nuevo_dia(monedas_base):
        for base in monedas_base:
            url = f"https://v6.exchangerate-api.com/v6/84b9e90f2667d75ca47bc21b/latest/{base}"
            response = requests.get(url)
            if response.status_code == 200:
                data = response.json()
                lista_moneda_base.append(base)
                lista_valor_conversion.append(data['conversion_rates'][target])
                lista_fecha.append(pd.to_datetime('today').strftime('%Y-%m-%d'))
            else:
                raise Exception(f"No se pudo obtener la tasa de cambio para {base} a {target}")

        # Crear un DataFrame con los nuevos datos
        df_tasa_cambio = pd.DataFrame(zip(lista_moneda_base, lista_valor_conversion, lista_fecha),
                                      columns=['moneda', f'valor_{target}', 'fecha'])

        # Guardar los datos en el archivo CSV para construir el histórico
        guardar_historico(df_tasa_cambio, ruta_historico)

        return df_tasa_cambio
    else:
        print("La tasa de cambio ya fue consultada hoy. No es necesario volver a ejecutarla.")
        return None

# Función para guardar o actualizar el archivo histórico
def guardar_historico(df_nuevo, ruta_historico):
    if os.path.exists(ruta_historico):
        # Si ya existe un archivo histórico, lo cargamos y lo concatenamos con el nuevo DataFrame
        df_nuevo.to_csv(ruta_historico, index=False, sep='|', encoding='utf-8-sig',mode='a',header=False)
    else:
        # Si no existe el archivo histórico, creamos uno nuevo
        df_nuevo.to_csv(ruta_historico, index=False,sep='|', encoding='utf-8-sig')
    #print(f"Datos guardados/actualizados en {ruta_historico}")
```
